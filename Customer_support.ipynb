```python
#droping the columns which have too many null value
df.drop(columns=["Customer Remarks","order_date_time","Customer_City","Product_category","Item_price","connected_handling_time"],inplace=True)
```

```python
df["Order_id"].fillna("unknown",inplace=True)
```

```python
df["Issue_reported at"]=pd.to_datetime(df["Issue_reported at"],errors="coerce")
df["issue_responded"]=pd.to_datetime(df["issue_responded"],errors="coerce")
df["responded_time"]=(df["Issue_reported at"]-df["issue_responded"]).dt.total_seconds()/60

#drop orginal column
df.drop(columns=["Issue_reported at","issue_responded"],inplace=True)


#fill missing value in responded_time column with median
df["responded_time"].fillna(df["responded_time"].median(),inplace=True)
```

```python
#checking still have missing value or not
df.isnull().sum()/len(df)*100
```

```python
#checking the duplicate values
df.duplicated().sum()
```

```python
# data type conversion
df['Survey_response_Date'] = pd.to_datetime(df['Survey_response_Date'])
df['Month'] = df['Survey_response_Date'].dt.month
df['Year'] = df['Survey_response_Date'].dt.year

df
```

```python
# This shows how your data is distributed across categories — for example, which channels (Chat, Call, Email) have the most records.
categorical_cols = ['channel_name', 'category', 'Sub-category', 
                    'Tenure Bucket', 'Agent Shift', 'Supervisor', 'Manager']

for col in categorical_cols:
    print(f"--- {col} ---")
    print(df[col].value_counts())
    print()
```

```python
#This bar chart helps you visualize which channels are used the most by customers.

plt.figure(figsize=(8,5))
sns.countplot(x='channel_name', data=df)
plt.title('Distribution of Channel Names')
plt.xticks(rotation=45)
plt.show()
```

```python
#This shows how customer satisfaction scores are spread out.
# If most scores are 4–5, it means high satisfaction; if centered around 2–3, customers are less happy.

sns.histplot(df['CSAT Score'],kde=True, bins=10, color='skyblue')
plt.title('Distribution of CSAT Scores')
plt.show()
```

```python
# This helps us compare average satisfaction across channels.
# Example: Customers may rate Chat support higher than Call support.

plt.figure(figsize=(8,5))
sns.barplot(x='channel_name', y='CSAT Score', data=df, palette='coolwarm')
plt.title('Average CSAT Score by Channel')
plt.xticks(rotation=45)
plt.show()
```

```python
# Boxplots show variation in CSAT for each category.

plt.figure(figsize=(10,5))
sns.boxplot(x='category', y='CSAT Score', data=df, palette='Set3')
plt.title('CSAT Score Distribution by Category')
plt.xticks(rotation=45)
plt.show()
```

```python
# Compares CSAT satisfaction between morning, evening, or night shifts. 

sns.barplot(x='Agent Shift', y='CSAT Score', data=df, palette='pastel')
plt.title('Average CSAT Score by Shift')
plt.show()
```

```python
#This combines three variables to see how performance differs by category and by shift.

sns.catplot(x='category', y='CSAT Score', hue='Agent Shift', kind='bar', data=df, height=5, aspect=2)
plt.title('CSAT Score by Category and Shift')
plt.xticks(rotation=45)
plt.show()
```

```python
# Heatmap shows which numerical feature are positively or negatively correlated

sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='Blues')
plt.title('Correlation Between Numeric Variables')
plt.show()
```

```python
# The box shows the range of normal data.
# The line inside the box is the median.
# The dots (outside the box) represent outliers — unusually high or low values.

plt.figure(figsize=(6,4))
sns.boxplot(x=df['CSAT Score'], color='lightcoral')
plt.title('Outlier Detection for CSAT Score')
plt.show()
```

```python
Q1 = df['CSAT Score'].quantile(0.25)
Q3 = df['CSAT Score'].quantile(0.75)
IQR = Q3 - Q1

lower_limit = Q1 - 1.5 * IQR
upper_limit = Q3 + 1.5 * IQR

outliers = df[(df['CSAT Score'] < lower_limit) | (df['CSAT Score'] > upper_limit)]
print("Number of Outliers in CSAT Score:", len(outliers))
```
